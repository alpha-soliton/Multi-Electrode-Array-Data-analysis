```python
A0_ibi_data = miPSdata_IBI_datas[0, 0][0]
A1_ibi_data = miPSdata_IBI_datas[0, 0][1]
A2_ibi_data = miPSdata_IBI_datas[0, 0][2]

B0_ibi_data = miPSdata_IBI_datas[0, 0][3]
B1_ibi_data = miPSdata_IBI_datas[0, 0][4]
B2_ibi_data = miPSdata_IBI_datas[0, 0][5]

C0_ibi_data = miPSdata_IBI_datas[0, 0][6]
C1_ibi_data = miPSdata_IBI_datas[0, 0][7]
C2_ibi_data = miPSdata_IBI_datas[0, 0][8]
```

```python
def compute_mean_ibi(ibi_data):
    tmp_sum = 0
    for i in range(len(ibi_data)):
        tmp_sum+=ibi_data[i][3]
    ibi_mean = tmp_sum/(len(ibi_data)-1)
    return ibi_mean
```

```python
A0_ibi_mean = compute_mean_ibi(A0_ibi_data)
A1_ibi_mean = compute_mean_ibi(A1_ibi_data)
A2_ibi_mean = compute_mean_ibi(A2_ibi_data)

B0_ibi_mean = compute_mean_ibi(B0_ibi_data)
B1_ibi_mean = compute_mean_ibi(B1_ibi_data)
B2_ibi_mean = compute_mean_ibi(B2_ibi_data)

C0_ibi_mean = compute_mean_ibi(C0_ibi_data)
C1_ibi_mean = compute_mean_ibi(C1_ibi_data)
C2_ibi_mean = compute_mean_ibi(C2_ibi_data)

A_ibi_mean = (A0_ibi_mean + A1_ibi_mean + A2_ibi_mean)/3
B_ibi_mean = (B0_ibi_mean + B1_ibi_mean + B2_ibi_mean)/3
C_ibi_mean = (C0_ibi_mean + C1_ibi_mean + C2_ibi_mean)/3
```
